```python
# Your y values
celsius_values = [25.6, 24.1, 26.7, 27.5, 28.3, 30.5, 32.8, 33.1]
print(celsius_values)
```

```python
# Your x values 
days = list(range(1, 9))
# Your y values
celsius_min = [19.6, 24.1, 26.7, 28.3, 27.5, 30.5, 32.8, 33.1]
celsius_max = [24.8, 28.9, 31.3, 33.0, 34.9, 35.6, 38.4, 39.2]
# Add Labels
plt.xlabel('Day')
plt.ylabel('Degrees Celsius')
plt.plot(days, celsius_min, label = "min") 
plt.plot(days, celsius_min, "ob") 
plt.plot(days, celsius_max, label = "max")
plt.plot(days, celsius_max, "or")
plt.legend()
plt.show()
```
